# Absa Bank Kenya: Rebranding Impact Analysis (2020-2025)
This notebook analyzes the transition from Barclays to Absa, focusing on brand investment impact on deposits and retail growth.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Connect to the integrated warehouse
engine = create_engine('postgresql://absa_admin:absa_password@localhost:5432/absa_warehouse')

# Load KPIs from marts
profitability = pd.read_sql('SELECT * FROM mart_profitability ORDER BY period', engine)
asset_quality = pd.read_sql('SELECT * FROM mart_asset_quality ORDER BY period', engine)
efficiency = pd.read_sql('SELECT * FROM mart_efficiency_ratio ORDER BY period', engine)

print("Loaded analytical marts successfully.")
```

## 1. Profitability Trends (ROE & NIM)
Post-rebranding efficiency and profitability.

```python
plt.figure(figsize=(12, 5))
plt.plot(profitability['period'], profitability['roe_percentage'], marker='o', label='ROE %')
plt.plot(profitability['period'], profitability['nim_percentage'], marker='s', label='NIM %')
plt.title('Absa Kenya: Profitability Metrics Post-Rebranding')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

## 2. Efficiency Ratio Analysis
Tracking the cost-to-income ratio as the bank stabilized after the brand transition.

```python
plt.figure(figsize=(12, 5))
plt.bar(efficiency['period'], efficiency['cost_to_income_ratio'], color='skyblue')
plt.title('Cost-to-Income Ratio Trend')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.axhline(y=50, color='red', linestyle='--', label='Target (50%)')
plt.legend()
plt.show()
```
